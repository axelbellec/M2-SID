## 1. Convergence de la loi hypergéométrique

On considère une variable $X^n_{N1,N2}$ de loi hypergéométrique ($min(N1, N2) \geq n$). Simuler pour différtentes valeurs de $n$, $N1$, $N2$ des k-échantillons de même loi que $X^n_{N1,N2}$. Comparer ces échantillions à ceux obtenus par les lois limites possibles (Binomiale et Poisson). On pourra comparer les deux lois empiriques.

```python
import numpy as np

def hypergeom(N1, N2, n):
    return np.random.hypergeometric(N1, N2, n)

def binomial(n, p):
    return np.random.binomial(n, p)

def poisson(lamb):
    return np.random.poisson(lamb)
```

__Convergence de la loi Hypergéométrique vers la loi Binomiale__

Si $N_1 \gg n$ et $N_2 \gg n$, alors la loi $\mathcal{H}(N_1, N_2, n) \backsimeq \mathcal{B}(n,p)$ ou $p = \frac{N_1}{N_1+N_2}$ .

```python
k = 100000

N1, N2, n = 2000, 1000, 40
p = N1/(N1+N2)

hypergeom_ech = [hypergeom(N1, N2, n) for _ in range(k)]
binomial_ech = [binomial(n, p) for _ in range(k)]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (10, 3)

num_bins = 50
f, axarr = plt.subplots(1, 2)
axarr[0].hist(hypergeom_ech, num_bins, normed=1, facecolor='green', alpha=0.5)
axarr[0].set_title('Hypergeometric distribution\n'+r'$n={0}$, $N1={1}$, $N2={2}$'.format(n, N1, N2))
axarr[0].set_ylabel('Probability')
axarr[0].set_xlabel('Smarts')
axarr[1].hist(binomial_ech, num_bins, normed=1, facecolor='blue', alpha=0.5)
axarr[1].set_title('Binomial distribution\n'+r'$n={0}$, $p={1}$'.format(n, round(p, 3)))
axarr[1].set_ylabel('Probability')
axarr[1].set_xlabel('Smarts')
plt.subplots_adjust(hspace=1.)
plt.show()
```

__Convergence de la loi Binomiale vers la loi de Poisson__

Si $n \geq 30$, $p \leq 0,1$ et $np < 15$, $\mathcal{B}(n,p) \backsimeq \mathcal{P}(np)$.

```python
k = 100000

n = 40 
p = N1/(N1+N2)

binomial_ech = [binomial(n, p) for _ in range(k)]
poisson_ech = [poisson(n*p) for _ in range(k)]
```

```python
num_bins = 50
f, axarr = plt.subplots(1, 2)
axarr[0].hist(binomial_ech, num_bins, normed=1, facecolor='green', alpha=0.5)
axarr[0].set_title('Binomial distribution\n'+r'$n={0}$, $p={1}$'.format(n, round(p, 3)))
axarr[0].set_ylabel('Probability')
axarr[0].set_xlabel('Smarts')
axarr[1].hist(poisson_ech, num_bins, normed=1, facecolor='blue', alpha=0.5)
axarr[1].set_title('Poisson distribution\n'+r'$\lambda={}$'.format(round(n*p, 3)))
axarr[1].set_ylabel('Probability')
axarr[1].set_xlabel('Smarts')
plt.subplots_adjust(hspace=1.)
plt.show()
```

## 2. Échantillonnage

On se place dans le cadre d’un plan d’échantillonnage simple. Ecrire un programme qui calcule $n$ et $c$ lorsque l’on se donne $(p_1, 1−\alpha)$ le risque fournisseur ; et $(p_2, \beta)$ le risque client $(p_1 < p_2)$. On se placera d’abord dans le cas hypergéométrique, puis on considèrera les approximations binomiales et de Poisson.

### Cas binomial


```python
from scipy.stats import binom

def bi_compute_p1(n, c, p1):
    return 1 - binom.cdf(c, n, p1)

def bi_compute_p2(n, c, p2):
    return binom.cdf(c, n, p2)
```

```python
def find_n_c_binom(p1, p2, alpha, beta):
    for n in range(1, 1000):
        for c in range(1, n):
            criteria_1 = bi_compute_p1(n, c, p1)
            criteria_2 = bi_compute_p2(n, c, p2)
            if criteria_1 <= alpha and criteria_2 <= beta:
                return n, c
    return None, None
```

```python
NQA = 0.001
NQL = 0.08
alpha = 0.1
beta = 0.2

n, c = find_n_c_binom(NQA, NQL, alpha, beta)
print('(n, c) = ({}, {})'.format(n, c))
```

### Cas poisson

```python
from scipy.stats import poisson

def poiss_compute_p1(c, plambda):
    return 1 - poisson.cdf(c, plambda)

def poiss_compute_p2(c, plambda):
    return poisson.cdf(c, plambda)
```

```python
def find_n_c_poiss(p1, p2, alpha, beta):
    for n in range(1, 1000):
        for c in range(1, n):
            criteria_1 = poiss_compute_p1(c, n*p1)
            criteria_2 = poiss_compute_p2(c, n*p2)
            if criteria_1 <= alpha and criteria_2 <= beta:
                return n, c
    return None, None
```

```python
NQA = 0.001
NQL = 0.08
alpha = 0.1
beta = 0.2

n, c = find_n_c_poiss(NQA, NQL, alpha, beta)
print('(n, c) = ({}, {})'.format(n, c))
```

### Cas hypergéométrique

```python
from scipy.stats import hypergeom


N = 1000

p1 = 0.05
p2 = 0.08

N1_provider = int(p1 * N)
N2_provider = N - N1_provider

N1_customer = int(p2 * N)
N2_customer = N - N1_customer

alpha = 0.1
beta = 0.2

for n in range(10, 300):
    found = False
    for c in range(1, n):
        provider_satisfied = 1 - hypergeom.cdf(c, N, N1_provider, n) < 1 - alpha
        customer_satisfied = hypergeom.cdf(c, N, N1_customer, n) <= beta
        if provider_satisfied and customer_satisfied:
            found = True
            break
    if found:
        break

print('(n, c) = ({}, {})'.format(n, c))
```
