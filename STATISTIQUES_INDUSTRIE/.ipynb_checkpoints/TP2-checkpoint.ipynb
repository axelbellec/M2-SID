## 1. Variance

On considère $n$ échantillons de longueur $k$ de la loi normale standard. Pour diverses valeurs de $k$ et $n$ comparer numériquement les estimateurs de l'écart-type vus en cours. Il faudra au préalable construire la table numérique des constantes $d_{2,k}$ et $c_{4,k}$ à l'aide d'une méthode de Monte Carlo.

```python
import numpy as np

def get_samples(k, n, mu=0, sigma=1):
    return [
        [np.random.normal(mu, sigma, k)] for ech in range(n)
    ]
```

On cherche à retrouver les valeurs de $d_2(k)$ quand $k \leq 10$ tout en sachant que $\hat{\sigma} \leadsto \frac{1}{d_2(k)}\frac{1}{n} \sum_{i=1}^n R_i$ ou $R_i = max_j(X_i^j) − min_j(X_i^j)$.

```python
def estimator_k_inf_10(samples):
    samples = list(samples)
    n = len(samples)
    return 1./n * np.sum([np.max(sample) - np.min(sample) for sample in samples])
```

```python
from collections import defaultdict
d2k_dict = defaultdict(lambda: defaultdict(lambda: defaultdict(float)))

for n in range(100, 2000, 100):
    for k in range(1, 10):
        samples = get_samples(k, n)
        d2k = estimator_k_inf_10(samples)
        d2k_dict[k][n] = d2k
```

```python
import pandas as pd

d2k_table = pd.DataFrame.from_dict(d2k_dict)
```

```python
d2k_table
```

On cherche à retrouver les valeurs de $c_4(k)$ quand $k \gt 10$ tout en sachant que $\hat{\sigma} \leadsto \frac{1}{c_4(k)}\frac{1}{n} \sum_{i=1}^n s_i$ ou $s_i = \sqrt{\frac{1}{n-1}\sum_{i=1}^{n}(x_i - \overline{x})^2}$.


```python
def estimator_k_sup_10(samples):
    samples = list(samples)
    n = len(samples)
    
    si = np.sqrt(1./(n-1) * [
        np.sum([(value - np.mean(sample))**2 for value in sample])
        for sample in samples
    ])
    
    return 1/n * si 
```

```python
c4k_dict = defaultdict(lambda: defaultdict(lambda: defaultdict(float)))

for n in range(100, 2000, 100):
    for k in range(10, 100, 10):
        samples = get_samples(k, n)
        c4k = estimator_k_sup_10(samples)
        c4k_dict[k][n] = c4k

c4k_table = pd.DataFrame.from_dict(c4k_dict)
```

## 2. Calcul des capabilités

> Ecrire des fonctions permettant le calcul des différentes capabilités. Les variables d'entrée sont $T_s$, $T_i$ et la matrice des observations $(k \times n)$.

```python
def compute_capability(matrix, t_i, t_s):
    
    m, s = np.mean(matrix), np.std(matrix)
    
    c_p = (t_s - t_i) / (6 * s)
    c_p_l = (m - t_i) / (3 * s)
    c_p_u = (t_s - m) / (3 * s)
    c_p_k = min(c_p_l, c_p_u)
    
    return c_p, c_p_l, c_p_u, c_p_k
```

On peut ensuite tester la fonction calculant la capabilité.

```python
t_i = -1
t_s = 1

matrix = np.random.randn(100, 10)

c_p, c_p_l, c_p_u, c_p_k = compute_capability(matrix, t_i, t_s)

print('c_p: {}'.format(round(c_p, 2)))
print('c_p_l: {}'.format(round(c_p_l, 2)))
print('c_p_u: {}'.format(round(c_p_u, 2)))
print('c_p_k: {}'.format(round(c_p_k, 2)))
```

## 3.

> On considère des suites de variables aléatoires $(X_{ij})_{i=1,...,n, j=1,...,k}$ toutes indépendantes de loi gaussienne de variance 1.  
On suppose que la moyenne de $X_{ij}$ vaut $\frac{1}{5} + \frac{1}{6} + \frac{1}{7} + ... + \frac{1}{5+n-1}$.  
Etudier l'évolution des différentes capabilités en fonction de $n$ (à $k$ fixé).

```python
from collections import namedtuple

Capability = namedtuple('capability', ['c_p', 'c_p_l', 'c_p_u', 'c_p_k'])

s, k, t_i, t_s = 1, 5, -1, 1

N = [i for i in range(10, 100, 10)]

capabilities = []

for n in N:
    m = 1 / (5 + n)
    X = [
        s * np.random.randn(1, k)[0] + (1 / (5 + j))
        for j in range(n)
    ]
    c_p, c_p_l, c_p_u, c_p_k = compute_capability(X, t_i, t_s)
    capabilities.append(Capability(c_p, c_p_l, c_p_u, c_p_k))
    
c_p_obs = [capability.c_p for capability in capabilities]
c_p_l_obs = [capability.c_p_l for capability in capabilities]
c_p_u_obs = [capability.c_p_u for capability in capabilities]
c_p_k_obs = [capability.c_p_k for capability in capabilities]

c_p_obs, c_p_l_obs, c_p_u_obs, c_p_k_obs
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = (10, 7)

fig, ax = plt.subplots()

n_groups = len(N)
index = np.arange(n_groups)

bar_width = 0.1
opacity = 0.4

plt.bar(
    index,
    c_p_obs,
    bar_width,
    alpha=opacity,
    color='b',
    label=r'$C_p$'
)

plt.bar(
    index + bar_width,
    c_p_l_obs,
    bar_width,
    alpha=opacity,
    color='r',
    label=r'$C_{pl}$'
)

plt.bar(
    index + 2 * bar_width,
    c_p_u_obs,
    bar_width,
    alpha=opacity,
    color='g',
    label=r'$C_{pu}$'
)

plt.bar(
    index + 3 * bar_width,
    c_p_k_obs,
    bar_width,
    alpha=opacity,
    color='y',
    label=r'$C_{pk}$'
)

plt.xlabel('N')
plt.ylabel('Capabilities')
plt.xticks(index + bar_width, N)
plt.legend(loc='lower right')
plt.grid()
```
